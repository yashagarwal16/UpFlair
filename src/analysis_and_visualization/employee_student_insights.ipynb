```python
import matplotlib.pyplot as plt

# Calculate department-wise average salary
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(avg_salary_by_dept.index, avg_salary_by_dept.values, color=['blue', 'green', 'red'])

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Department-wise Average Salary")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv("/UPFLAIR/notebook/data/employee.csv")

# Count employees in each department
dept_counts = df["Department"].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(dept_counts.index, dept_counts.values, color=['blue', 'green', 'red'])

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in Each Department")

# Show the plot
plt.show()
```

```python
# Filter employees with salary > 60000
high_salary_employees = df[df["Salary"] > 60000]

# Print the result
print(high_salary_employees)
```

```python
# Calculate average salary per department
avg_salary_by_dept = df.groupby("Department")["Salary"].mean()

# Plot the line graph
plt.figure(figsize=(8, 5))
plt.plot(avg_salary_by_dept.index, avg_salary_by_dept.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Salary Trend by Department")

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/UPFLAIR/notebook/data/student.csv")

# Calculate mean of each subject
mean_scores = df.iloc[:, 1:].mean().to_dict()

# Print mean scores
print("Mean Scores:", mean_scores)
```

```python
# Calculate average score per student
df["Average Score"] = df.iloc[:, 1:].mean(axis=1)

# Print the updated DataFrame
print(df[["Student Name", "Average Score"]])
```

```python
# Calculate total score for each student
df["Total Score"] = df.iloc[:, 1:5].sum(axis=1)

# Find the student with the highest score
top_student = df.loc[df["Total Score"].idxmax(), "Student Name"]
print("Top Student:", top_student)
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(df["Math"], df["Science"], color='blue', label="Math vs Science")

# Add labels
plt.xlabel("Math Scores")
plt.ylabel("Science Scores")
plt.title("Scatter Plot: Math vs Science Scores")
plt.legend()
plt.show()
```

```python
# Standardizing: (X - mean) / std
standardized_scores = (df.iloc[:, 1:5] - df.iloc[:, 1:5].mean()) / df.iloc[:, 1:5].std()

# Print standardized scores
print(standardized_scores)
```

```python
plt.figure(figsize=(6, 4))
plt.bar(df["Student Name"], df["Total Score"], color=['blue', 'green', 'red', 'purple'])

# Add labels
plt.xlabel("Students")
plt.ylabel("Total Score")
plt.title("Total Scores of Students")
plt.show()
```
